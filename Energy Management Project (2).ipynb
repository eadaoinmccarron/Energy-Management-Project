## <font color='Navy'>Introduction</font>

Analysing and improving the thermal performance of buildings has become a critical part of engineering in our world today. A thermal efficient building decreases operational costs and increases productivity. Learning how to predict the thermal efficiency of a building can help us create the most optimal solution for each building. This includes the dimensions of the building, the HVAC systems, the positioning of windows and doors, quality and width of material and many other aspects. 

This project gave us the chance to experiment with various details in regards to the thermal performance of an appartment. In this report, we have outlined the hypothesis used and the assumptions that were made. We have chosen a dwelling with two rooms, a living room/kitchen and a bathroom. We have set the location as Lyon, France and used the weather data accordingly, analysing the effect of the temperatures in both winter and in summer. The kitchen/living room is 25m2 and the bathroom is 10m2. In this project, we used Python as a tool to examine and calculate the thermal properties of this appartment. 

After completing the initial evaluation of the house, we have implemented changes to our design to optimise the thermal performance. These changes included substituting the initial materials used to those of better quality, implementing a PID controller, and changing the widths of materials used. In doing so, we can compare the results and find the optimal solution for an efficient and comfortable dwelling.
